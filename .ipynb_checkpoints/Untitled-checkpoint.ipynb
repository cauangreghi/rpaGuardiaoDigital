```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import time
import pandas as pd
import pyautogui
import shutil
import os
import jpype

cwd = os.getcwd()
path = cwd + "\\"
global df1
i = 0

caminhos = [os.path.join(path, nome) for nome in os.listdir(path)]
arquivos = [arq for arq in caminhos if os.path.isfile(arq)]
pdfs = [arq for arq in arquivos if arq.endswith(".pdf")]
pdfs = list(map(lambda x: x.replace(path, ""), pdfs))

servico = Service(ChromeDriverManager().install())
options = webdriver.ChromeOptions()
prefs = {'download.prompt_for_download': True,
         'download.directory_upgrade': True,
         'plugins.always_open_pdf_externally' : False,
         "profile.default_content_setting_values.notifications" : 2
        }

options.add_experimental_option('prefs', prefs)
navegador = webdriver.Chrome(service=servico, options=options)

navegador.get("https://cedoc.guardiao.digital/#/login/company")
navegador.maximize_window()

with open("venv.txt", "r") as arquivo:
    texto = arquivo.readlines()
usu = texto[0].split("/n")[0]
pas = texto[0].split("/n")[1]

time.sleep(5)
navegador.find_element("xpath", '/html/body/app-root/gd-login/div/div[3]/gd-company/div/div/div/div/form/div[1]/div/input').send_keys("rni")
navegador.find_element("xpath",'/html/body/app-root/gd-login/div/div[3]/gd-company/div/div/div/div/form/button').click()
time.sleep(2)
navegador.find_element("xpath",'/html/body/app-root/gd-login/div/div[3]/gd-user-password/div/form/div/div[1]/input').send_keys(usu)
navegador.find_element("xpath",'/html/body/app-root/gd-login/div/div[3]/gd-user-password/div/form/div/div[2]/input').send_keys(pas)
time.sleep(1)
navegador.find_element("xpath",'/html/body/app-root/gd-login/div/div[3]/gd-user-password/div/form/button').click()
time.sleep(1)
navegador.find_element("xpath",'/html/body/app-root/gd-login/div/div[1]/button').click()

time.sleep(6)
ActionChains(navegador)\
    .send_keys(Keys.ESCAPE)\
    .send_keys(Keys.ESCAPE)\
    .perform()
time.sleep(2)

def rpaGed():
    global df1
    
    navegador.find_element("xpath",'/html/body/app-root/gd-ged/main/section/div/gd-pesquisa/div/div[1]/div[2]/gd-simple-search/div/div[1]/div/div[2]/input').send_keys(pdfs[i].split(".")[0])
    print(pdfs[i])
    ActionChains(navegador)\
        .send_keys(Keys.ENTER)\
        .perform()
    time.sleep(3)
    try:
        clickable = navegador.find_element("xpath","// span[contains(text(),\'MALA DIRETA')]")
        ActionChains(navegador)\
            .double_click(clickable)\
            .perform()

        time.sleep(3)

        navegador.find_element("xpath","// span[contains(text(),\'DOCUMENTACAO COMPLETA')]").click()
        iframe = navegador.find_element("xpath","// span[contains(text(),\'AR/Telegrama/Protocolo ')]")
        ActionChains(navegador)\
            .scroll_to_element(iframe)\
            .click(iframe)\
            .perform()
        time.sleep(2)
        navegador.find_element_by_xpath("/html/body/app-root/gd-ged/main/section/div/gd-documentos/div/gd-document/div/div[3]/gd-document-files-folders/div/gd-box/div/gd-button[2]/button").click()
        time.sleep(4)
        navegador.find_element_by_xpath("/html/body/app-root/smart-modal/div[1]/div/div/div[2]/gd-document-insert-file/form/div[2]/div").click()
        time.sleep(4)
        pyautogui.write(path+pdfs[i])
        time.sleep(4)
        pyautogui.press('enter')
        time.sleep(4)
        #save
        navegador.find_element_by_xpath("/html/body/app-root/smart-modal/div[1]/div/div/div[2]/gd-document-insert-file/div[2]/gd-button[3]").click()

        ActionChains(navegador)\
            .key_down(Keys.LEFT_CONTROL)\
            .send_keys("q")\
            .send_keys("q")\
            .key_up(Keys.LEFT_CONTROL)\
            .perform()
        
        dfCabecalho = pd.DataFrame(
            {'Contrato': [empre],
             'Situação': "Importada a notificação"
                }
            )

        if i == 0:
            df1 = dfCabecalho
        else:
            df1 = pd.concat([df1, dfCabecalho])
        return df1

    except:
        navegador.find_element("xpath", "/html/body/app-root/gd-ged/main/div/div[2]").click()
        time.sleep(2)
        iframe = navegador.find_element("xpath","// div[contains(text(),\' MALA DIRETA ')]")
        ActionChains(navegador)\
            .scroll_to_element(iframe)\
            .double_click(iframe)\
            .perform()
        pdfs2 = pdfs[i].split("-")
        navegador.find_element("xpath", "/html/body/app-root/gd-ged/main/section/div/gd-documentos/div/gd-document/div/div[1]/div/gd-document-form/form/gd-box-form[4]/div/gd-smart-select/ng-select/div/div/div[2]/input").send_keys(pdfs2[0])
        ActionChains(navegador)\
            .send_keys(Keys.ENTER)\
            .perform()
        navegador.find_element("xpath", "/html/body/app-root/gd-ged/main/section/div/gd-documentos/div/gd-document/div/div[1]/div/gd-document-form/form/gd-document-form-dynamic/gd-box-form[2]/div/gd-input-text/div/div/input").send_keys(pdfs2[1])
        navegador.find_element("xpath", "/html/body/app-root/gd-ged/main/section/div/gd-documentos/div/gd-document/div/div[1]/gd-box/div/gd-button[2]/button").click()
        time.sleep(3)
        navegador.find_element_by_xpath("/html/body/app-root/gd-ged/main/section/div/gd-documentos/div/gd-document/div/div[3]/gd-document-files-folders/div/gd-box/div/gd-button[2]/button").click()
        time.sleep(3)
        navegador.find_element_by_xpath("/html/body/app-root/smart-modal/div[1]/div/div/div[2]/gd-document-insert-file/form/div[2]/div/label").click()
        time.sleep(3)
        pyautogui.write(path+pdfs[i])
        time.sleep(1)
        pyautogui.press('enter')
        time.sleep(3)
        #save
        navegador.find_element_by_xpath("/html/body/app-root/smart-modal/div[1]/div/div/div[2]/gd-document-insert-file/div[2]/gd-button[3]/button").click()
        dfCabecalho = pd.DataFrame(
            {'Contrato': [empre],
             'Situação': "Criado a pasta e importado a notificação"
                }
            )

        if i == 0:
            df1 = dfCabecalho
        else:
            df1 = pd.concat([df1, dfCabecalho])

        return df1
    
def exportExcel(writer):
    writer = pd.ExcelWriter("LOG-GED.xlsx")
    df1.to_excel(writer, index=False)
    writer.close()

while i < 2:
    try:
        rpaGed()
        print(f"Documento {i + 1} de {len(pdfs)} concluído")
        i += 1
    except:
        print(f"Processo parou no documento {i + 1} de {len(pdfs)}")
        break

exportExcel(df1)
input("=============== PROCESSO FINALIZADO ===============")
```
